### Вопрос 16
Винни-Пух в течение 150 дней фиксировал изменения количества пчёл в улье. Он уверен, что полученные наблюдения являются выборкой независимых одинаково распределённых нормальных случайных величин. Оказалось, что среднее количество пчёл равно 25000, а выборочная дисперсия равна 1300. Постройте 95%-ый доверительный интервал для математического ожидания количества пчёл в улье и выпишите в ответ его нижнюю границу, округлённую до сотых.

```python
import numpy as np
```

```python
mean = 25000
var = 1300
n = 150

lower = mean - 1.96 * np.sqrt(var) / np.sqrt(n)
```

```python
round(lower, 2)
```

### Вопрос 31
Оцените линейную регрессию, уравнение которой имеет следующий вид:<br>
$Global\_Sales_i = w_0 + w_1 * ln(1 + NA\_Sales_i)$,<br>
где нижние индексы обозначают значения соответствующих переменных для наблюдения i.<br>
В ответ выпишите оценку коэффициента при свободном члене.

```python
# Сделаем новую переменную со значением ln(1+NA_Sales).
df['ln_NA_Sales'] = np.log(1 + df['NA_Sales'])
df['ln_NA_Sales']
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
X = df['ln_NA_Sales'].to_numpy().reshape(len(df['ln_NA_Sales']), 1)
y = df['Global_Sales'].to_numpy().reshape(len(df['Global_Sales']), 1)
```

```python
model.fit(X, y)
```

```python
model.intercept_
```
